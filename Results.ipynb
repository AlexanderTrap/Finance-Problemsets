```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = 'FB', 'GLD', 'GNUS', 'K', 'NTDOY'
sizes = [5.3, 53.3, 0.1, 18.4,22.8]
explode = (0, 0, 0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.savefig("pie_diagram.png",bbox_inches='tight')
plt.show()
```

```python
from datetime import datetime
today=datetime.today().strftime('%Y-%m-%d')
```

```python
tickers_list = ['NTDOY','GNUS','FB','K','GLD']
tickers_list=sorted(tickers_list)
tickers_list
```

Weight Distriution

```python
pf=np.array([5.3,53.3,0.1,18.4,22.8])/100
```

```python
# Get the data for the stock AAPL
data = yf.download(tickers_list,'2020-06-15','2020-06-30')
(data['Adj Close']).plot(figsize=(10,6),linestyle='-')
plt.title('Stocks', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Adj Close',fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.legend()
# plt.savefig("stocks_pf.png",bbox_inches='tight')
plt.show()
```

```python
d=yf.download(tickers_list,'2020-06-15','2020-06-30',interval='1d')
data=d['Adj Close'].dropna()
```

```python
data
```

```python
n=(data/data.iloc[0]-1)*100
n
```

```python
n.plot(figsize=(15,6),linestyle='-')
plt.title('Return', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Return (%)',fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
# plt.savefig("n_5.png",bbox_inches='tight')
plt.show()
```

```python
df3=(data/data.iloc[0]-1)*pf*100
df3
```

```python
df3.plot(figsize=(15,6),linestyle='-')
# plt.title(, fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Return (%)',fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
# plt.savefig("pf_5.png",bbox_inches='tight')
plt.show()
```

```python
DF3=df3.sum(axis=1)
```

```python
DF3.plot(figsize=(10,6),linestyle='-')
plt.title('Total Portfolio return', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Return (%)',fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
# plt.savefig("pf.png",bbox_inches='tight')
plt.show()
```
